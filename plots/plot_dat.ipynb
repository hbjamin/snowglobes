```python
import matplotlib.pyplot as plt

def plot_data_file(filename):
    x_vals = []
    y_vals = []
    with open(filename, 'r') as f:
        for line in f:
            # skip empty lines and lines with separators like dashes or totals
            line = line.strip()
            if not line or line.startswith('-') or 'Total' in line:
                continue
            parts = line.split()
            if len(parts) != 2:
                continue
            try:
                x = float(parts[0])
                y = float(parts[1])
                x_vals.append(x)
                y_vals.append(y)
            except ValueError:
                # skip lines that don't parse as floats
                continue

    plt.figure(figsize=(8,5))
    plt.plot(x_vals, y_vals, marker='o', linestyle='-', markersize=3)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f'Data from {filename}')
    plt.grid(True)
    plt.show()

# Example usage:
plot_data_file('/nfs/disk1/users/bharris/SN_stuff/snowglobes/out/stpi_nue_e_d2O_events.dat')
```
